```python
df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],'B':['B8','B9','B10','B11'],'C':['C8','C9','C10','C11'],'D':['D8','D9','D10','D11']},index=[8,9,10,11])
df3
```

## Merge

```python
left = pd.DataFrame({'Key':['K0','K8','K2','K3'],'B':['B0','B1','B2','B3'],'C':['C0','C1','C2','C3']},index=[0,1,2,3])
left
```

```python
right = pd.DataFrame({'Key':['K0','K1','K2','K3'],'B':['B4','B5','B6','B7'],'C':['C4','C5','C6','C7']},index=[0,1,2,3])
right
```

```python
merge = pd.merge(left,right,how='inner',on='Key') # Inner Join with common join column 'Key'
merge
```

```python
df1 = pd.DataFrame({'Key1':['K0','K8','K2','K3'],'B':['B0','B1','B2','B3'],'C':['C0','C1','C2','C3']},index=[0,1,2,3])
df1
```

```python
df2 = pd.DataFrame({'Key2':['K0','K1','K2','K3'],'B':['B4','B5','B6','B7'],'C':['C4','C5','C6','C7']},index=[0,1,2,3])
df2
```

```python
merge1 = pd.merge(df1,df2,how='inner',left_on='Key1',right_on='Key2') 
# Inner Join with  join point without common column, 'Key1' on left and 'Key2' on right
merge1
```

```python
merge1 = pd.merge(df1,df2,how='left',left_on='Key1',right_on='Key2') 
# Left Join with  join point without common column, 'Key1' on left and 'Key2' on right
merge1
```

```python
merge2 = pd.merge(df1,df2,how='right',left_on='Key1',right_on='Key2') 
# Right Join with  join point without common column, 'Key1' on left and 'Key2' on right
merge2
```

### 11. Compare by how much percent one beer has more alcohol content than another beer

```python
def compare_alcohol_percent():
    beer_1 = input("Enter beer name: ")
    beer_2 = input("Enter beer name: ")
    
    beer1 = df1[df1['name']==beer_1]['alcohol'].sum()
    print("{} alcohol content is {}".format(beer_1,beer1))
    beer2 = df1[df1['name']==beer_2]['alcohol'].sum()
    print("{} alcohol content is {}".format(beer_2,beer2))
    
    if beer2 > beer1:  
        alcohol_percent = beer1/beer2
        print(alcohol_percent)
        percent = (1-alcohol_percent)*100
        print("{} has {}% more alcohol than {}".format(beer_2,percent,beer_1))
    else:
        alcohol_percent = beer2/beer1
        print(alcohol_percent)
        percent = (1-alcohol_percent)*100
        print("{} has {}% more alcohol than {}".format(beer_1,percent,beer_2))
    
```
